# Data Cleaning

**Data cleaning** is the process of identifying and correcting errors or inconsistencies in a dataset. This includes handling missing values, removing duplicates, correcting data types, detecting and addressing outliers, and normalizing or encoding features. Proper data cleaning ensures the quality and accuracy of data before analysis or modeling.


```python
zero_count_per_column = (df == 0).sum()
print("Number of zeros in each column:")
print(zero_count_per_column)
```

```python
correlation = df.corr()
# Plot heatmap
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.show()
```

```python
 # Replacing the zero values with no zeros values mean( according to their columns)
df['Insulin'] = df['Insulin'].replace(0, df[df['Insulin'] != 0]['Insulin'].mean())
df['Glucose'] = df['Glucose'].replace(0, df['Glucose'][df['Glucose'] != 0].mean())
df['BloodPressure'] = df['BloodPressure'].replace(0, df['BloodPressure'][df['BloodPressure'] != 0].mean())
df['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'][df['SkinThickness'] != 0].mean())
df['BMI'] = df['BMI'].replace(0, df['BMI'][df['BMI'] != 0].mean())
df
```

# Data Preprocessing

**Data preprocessing** is the process of transforming raw data into a clean and structured format for machine learning. It includes tasks like splitting the data into training and testing sets, handling missing values, scaling features, encoding categorical variables, and addressing class imbalance to ensure the model performs effectively.


```python
# Define the target variable (Outcome)
X = np.asarray(df.drop(columns='Outcome'))
Y = np.asarray(df['Outcome'])
#Standardizing or normalizing the data ensures that features have similar scales
scaler = StandardScaler()
X=scaler.fit_transform(X)
X
```

```python
# Split data into training and testing sets (80% train, 20% test)
x_train, x_test, y_train,y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Plotting the scatter plot of BMI vs Age, coloring by Outcome (0: non-diabetic, 1: diabetica
```

# Model: SVM (Support Vector Machine)

**Support Vector Machine (SVM)** is a supervised learning algorithm used primarily for classification tasks. It works by finding the hyperplane that best separates data into different classes, aiming to maximize the margin between these classes.

### Key Points:
- **Hyperplane**: The boundary that divides classes.
- **Support Vectors**: Data points closest to the hyperplane, critical in defining it.
- **Kernel Trick**: Enables SVM to handle non-linear data by mapping it to a higher-dimensional space.
  
SVM works well with:
- High-dimensional data.
- Clear margins of separation.

```python
# Plotting the scatter plot of BMI vs Age, coloring by Outcome (0: non-diabetic, 1: diabetic)
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='BMI', y='Age', hue='Outcome', palette={0: 'blue', 1: 'red'}, markers=["o", "X"])

# Add labels and title
plt.title('Scatter Plot of BMI vs Age (Diabetes Prediction)')
plt.xlabel('BMI')
plt.ylabel('Age')

# Show the plot with a legend
plt.legend(title='Outcome', loc='upper right')
plt.show()
```

```python
#the dataset is non-linear therefor kernel -rbf
model = SVC(kernel='rbf')
model.fit(x_train, y_train)
```

```python
pre=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,pre)
print(f'Model accuracy:{accuracy * 100:.2f}%')
```

```python
# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test,pre))
# Classification Report
print("Classification Report:")
print(classification_report(y_test,pre))
```

```python
def predict_diabetes(input_data):
    # Preprocess the input data (same preprocessing as during training)
    input_scaled = scaler.transform([input_data])  # Scale the input features using the same scaler
    
    # Predict using the trained SVM model
    prediction = model.predict(input_scaled)
    
    # Convert the prediction into a more understandable result
    print(prediction)
    if prediction == 1:
        return "Diabetic"
    else:
        return "Healthy"


input_data = [9	,171	,110	,24	,240	,45.4	,0.721	,54]
# input_data.append(float(input("Enter Pregnancies: ")))
# input_data.append(float(input("Enter Glucose Level: ")))
# input_data.append(float(input("Enter BloodPressure Level: ")))
# input_data.append(float(input("Enter SkinThickness: ")))
# input_data.append(float(input("Enter Insulin Level: ")))
# input_data.append(float(input("Enter BMI: ")))
# input_data.append(float(input("Enter DiabetesPedigreeFunction: ")))
# input_data.append(float(input("Enter Age: ")))

input_data=input_data
# Call the function to make a prediction
result = predict_diabetes(input_data)
print(f"The model predicts: {result}")
```
